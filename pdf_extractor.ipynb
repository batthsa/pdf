# PDF Page Extractor
Upload a PDF and extract specific pages.

```python
from google.colab import files
uploaded = files.upload()
```

```python
from PyPDF2 import PdfReader, PdfWriter

def extract_pages(input_pdf, output_pdf, pages):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    for p in pages:
        writer.add_page(reader.pages[p - 1])

    with open(output_pdf, 'wb') as f:
        writer.write(f)
    print('Created:', output_pdf)
```

```python
# Example usage
input_pdf = list(uploaded.keys())[0]
output_pdf = 'output_pages_4_to_9.pdf'
pages_to_extract = [4,5,6,7,8,9]

extract_pages(input_pdf, output_pdf, pages_to_extract)
```
